```python
df = pd.read_csv('TITANIC.csv')
```

QUESTION 1: What is the median fare paid by passenger?

```python
median_value = df['Fare'].median()
print(median_value)
```

QUESTION 2: What is the mean age of male passengers

```python
mean_value = df['PassengerId'].mean()
male_df = df[df['Sex'] == 'male']
mean_age_of_males = male_df['Age'].mean()
print(mean_age_of_males)
```

QUESTION 3: What is the mode of the number of siblings/spouses aboard on the titanic?

```python
mode_value = df['SibSp'].mode().iloc[0]
print(mode_value)
```

QESTION4: What was the range of ticket prices?

```python
max_value = df['Fare'].max()
min_value = df['Fare'].min()
data_range = max_value - min_value
print(f"Range of 'Fare' column: {data_range}")
```

QUESTION5: How much did the cheapest ticket cost?

```python
cheapest_ticket_cost = df['Fare'].min()
print(f"Cheapest ticket cost: {cheapest_ticket_cost}")
```

QUESION6: Compute the correlation between Sex and Survival.

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
correlation = df['Sex'].corr(df['Survived'])
print(f"Correlation between 'Sex' and 'Survived': {correlation}")
```

QUESTION 7: Find the variance and the standard deviation of the passenger class.

```python
variance = df['Pclass'].var()
std_deviation = df['Pclass'].std()
print(f"Variance of 'Passenger class': {variance}")
print(f"Standard Deviation of Passenger class: {std_deviation}")
```

## **Data Visualization Techniques**



The techniques that would be best to implement to find outliers in the dataset are:

a. Box Plot: The box plot will be best because it provides a visual summary of the distribution
of the dataset by displaying the median, quartiles and any potential outliers.

b. Scatter Plots: This displays individual data points on a two-dimensional plane, which
makes it easy to visualize relationships between two variables.

c. Histogram: This technique will also be suitable since it displays the distribution of a single
variable by dividing it into bins adn counting the number of data points in each bin.

```python
sns.set(style = "whitegrid")
#Box Plot (PCLASS VS AGE)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.boxplot(x='Pclass', y='Age', data=df, ax=axes[0])
axes[0].set_title('Box Plot (Pclass vs. Age)')
#Scatter Plot (FARE VS AGE)
sns.scatterplot(x='Fare', y='Age', data=df, ax=axes[1])
axes[1].set_title('Scatter Plot (Fare vs. Age)')
#Histogram (AGE)
sns.histplot(df['Age'], bins=20, kde=True, ax=axes[2])
axes[2].set_title('Histogram (Age)')
```
