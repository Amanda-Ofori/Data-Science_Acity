Outliers are values within a dataset that vary greatly from the others—they're either much larger, or significantly smaller. Outliers may indicate variabilities in a measurement, experimental errors, or a novelty.(CareerFoundry.com). Explain three common ways of detecting outliers in the given dataset.

- Interquartile Range (IQR) Method: This method is also used to identify outliers and is not affected by extreme values. The IQR describes the middle 50% of values when ordered from lowest to highest. To find the interquartile range (IQR), ​first find the median (middle value) of the lower and upper half of the data. These values are quartile 1 (Q1) and quartile 3 (Q3). The IQR is the difference between Q3 and Q1(Khan Academy.com).

- Data Visualization : With box plot, use the box plot, or the box and whisker plot, to explore the dataset and visualize the presence of outliers. The points that lie beyond the whiskers are detected as outliers. You can generate box plots in Seaborn using the boxplot function(freeCodeCamp.com).To use a box plot to detect outliers, you need to look for data points that are outside the whiskers, which are the lines that extend from the box to the minimum and maximum values. These data points are considered outliers, as they are more than 1.5 times the IQR away from the box(LinkedIn

- This method is used to identify outliers in a dataset by measuring the amount of stadard deviations the data point is away from the mean.For example, a z-score of 2.5 indicates that the data point is 2.5 standard deviation away from the mean. Usually z-score =3 is considered as a cut-off value to set the limit. Therefore, any z-score greater than +3 or less than -3 is considered as outlier which is pretty much similar to standard deviation method(TowardsDataScience.com). It works by:

i. Calculating the mean(μ) and standard deviation(σ) of the dataset.

ii. For each data point, the Z-Score is computed using the formula: Z = (X - μ) / σ, where X is the data point.

iii. Normally, a threshold value which is normally set to 3 or -3, is used. Data points with Z-Scores that exceed the threshold are considered outliers.
